#CHATBOT WITH RULE-BASED RESPONSES:

Build a simple chatbot that responds to user inputs based on
predefined rules. Use if-else statements or pattern matching
techniques to identify user queries and provide appropriate
responses. This will give you a basic understanding of natural
language processing and conversation flow

```python
#imports
import re
import random

#Class Definition
class RuleBasedChatbot:
    def __init__(self):
        self.responses = {
            r'\b(hi|hello|hey)\b': ["Hello! How can I assist you today?", "Hi there! How can I help you?"],
            r'\b(how are you|how\'s it going)\b': "I'm just a chatbot, but thank you for asking!",
            r'\b(what\'s your name|who are you)\b': "I'm a rule-based chatbot programmed to assist you!",
            r'\b(what can you do)\b': "I can provide information, answer questions, or just chat with you!",
            r'\b(help)\b': "Sure! You can ask me anything, and I'll do my best to assist you.",
            r'\b(weather)\b': "I'm sorry, I don't have access to weather information.",
            r'\b(goodbye|bye)\b': "Goodbye! Have a great day!",
            r'\b(my name is (\w+))\b': "Nice to meet you, {}!",
            r'\b(what is my name)\b': "I'm sorry, I don't know your name.",
            r'\b(thank you|thanks)\b': "You're welcome! Feel free to ask anything else.",
            r'\b(sorry|apologies)\b': "No problem! Let me know how I can help.",
            r'\b(exit|quit)\b': "Goodbye! Have a great day!"
        }

#Process Input Method
    def process_input(self, user_input):
        user_input = user_input.lower()
        for pattern, response in self.responses.items():
            match = re.search(pattern, user_input)
            if match:
                if isinstance(response, list):
                    return random.choice(response)
                elif '{}' in response:
                    return response.format(match.group(2) if match.group(2) else '')
                else:
                    return response
        return "I'm sorry, I didn't understand that."

#Start Chat Method
    def start_chat(self):
        print("Welcome to the Rule-Based Chatbot!")
        print("You can start chatting. Type 'exit' or 'quit' to end the conversation.")

#Error Handling
        while True:
            try:
                user_input = input("You: ")
                if user_input.lower() in ['exit', 'quit']:
                    print("Chatbot: Goodbye! Have a great day!")
                    break
                response = self.process_input(user_input)
                print("Chatbot:", response)
            except KeyboardInterrupt:
                print("\nChatbot: Oops! Looks like you interrupted the conversation. Goodbye!")
                break
            except Exception as e:
                print("Chatbot: An error occurred:", e)

#Main Execution
if __name__ == "__main__":
    chatbot = RuleBasedChatbot()
    chatbot.start_chat()
```
